```python
import numpy as np
import matplotlib.pyplot as plt

def get_positional_encoding(pos, d_model):
    angle_rads = np.arange(d_model)[np.newaxis, :] // 2 * 2 / d_model
    angle_rads = pos / np.power(10000, angle_rads)

    # apply sine to even indices in the array; 2i
    angle_rads[:, 0::2] = np.sin(angle_rads[:, 0::2])

    # apply cosine to odd indices in the array; 2i+1
    angle_rads[:, 1::2] = np.cos(angle_rads[:, 1::2])

    return angle_rads

# Increase the number of positions
positions = np.arange(100)[:, np.newaxis]
d_model = 4

positional_encoding = get_positional_encoding(positions, d_model)

```

```python

# Plot the positional encodings
plt.figure(figsize=(10, 6))
plt.plot(positional_encoding[:, 0], label="dim 0 (sin)")
plt.plot(positional_encoding[:, 1], label="dim 1 (cos)")
plt.plot(positional_encoding[:, 2], label="dim 2 (sin)")
plt.plot(positional_encoding[:, 3], label="dim 3 (cos)")
plt.xlabel('Position in Sequence')
plt.ylabel('Positional Encoding Value')
plt.title('Positional Encodings Over Sequence Positions')
plt.legend()
plt.show()
```
